```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('comparison_analytical_vs_experimental_results.csv')
df.tail()
```

```python
plt.plot(df['N'], df['Analytical probability of Alice wins'], label='Theory', marker = 'o', linestyle = '')
plt.plot(df['N'], df['Experimental probability of Alice wins'], label='Experiment', marker='x', linestyle ='')
plt.xlabel('Number of Coin Tosses in Each Game')
plt.ylabel("Alice's Probability of Winning")
plt.grid()
plt.legend()
plt.show()
```
